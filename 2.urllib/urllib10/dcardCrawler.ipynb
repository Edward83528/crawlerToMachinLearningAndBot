```python
#coding:utf-8
import urllib.request
import json
import codecs
import sys
import argparse as ap
import time
import datetime
import lxml.html
from urllib.parse import quote
from random import randint

def argParse():
    parser=ap.ArgumentParser(description='Dcard Crawler')
    parser.add_argument("category", help="Serch Category,EX:food:")
    return parser.parse_args()

args=argParse()
category = quote(args.category)

#category = quote('funny')


def start_requests():
    urls = []
    api = 'https://www.dcard.tw/f/'+category
    urls.append(api)
    for url in urls:
        #print (url)
        parseLtnDcard(url)
        time.sleep(0.2)


def parseLtnDcard(uri):
    print(uri)
    req = urllib.request.Request(
        uri, 
        data=None, 
        headers={
            'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/35.0.1916.47 Safari/537.36'
        }
    )
    handle = urllib.request.urlopen(req)
    encoding = handle.headers.get_content_charset()
    html_data =  handle.read().decode(encoding)
    #print(html_data)
    selector = lxml.html.document_fromstring(html_data)
    newslist = selector.xpath('//*[@class="sc-1v1d5rx-3 kPUUNB"]')
    #print(len(newslist))
    for i in range(len(newslist)):
        #print(newslist[i]);
        strTitle = ''
        strUrl = ''
        str_idx = str(i+1)
        str_xpath = '//*[@class="sc-1v1d5rx-3 kPUUNB"]//span//text()'
        titleList = selector.xpath(str_xpath)[i]
        strTitle = " ".join(titleList)
        print(str_idx+":"+strTitle,end='\n')
        
        str_xpath = '//*[@class="sc-1v1d5rx-3 kPUUNB"]/@href'
        urlList = selector.xpath(str_xpath)[i]
        strUrl = ''.join(urlList)
        strUrl = strUrl
        print(strUrl)
        print('='*30,end='\n')
        if len(strTitle)>1:
            items.append({
                "title": strTitle,
                "link":strUrl,
                })
    handle.close()

if __name__ == '__main__':
    items = []
    start_requests()
    row_json = json.dumps(items, ensure_ascii=False)
    file = codecs.open(urllib.parse.unquote(keyword)+'.json', 'w', encoding='utf-8')
    #file = codecs.open('out.json', 'w', encoding='utf-8')
    file.write(row_json)
    file.close()

    print("Done")
```
